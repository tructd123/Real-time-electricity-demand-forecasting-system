```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from datetime import datetime

def create_individual_fuel_type_visualizations(df: pd.DataFrame, save_plots: bool = True, output_dir: str = "fuel_type_plots"):
    """
    Create individual visualizations for daily fuel type data and save each as a separate PNG file.

    Args:
        df: DataFrame containing daily fuel type data
        save_plots: Whether to save plots to files
        output_dir: Directory to save the plot files
    """
    print("\n Creating individual fuel type visualizations...")

    # Create output directory if it doesn't exist
    if save_plots:
        os.makedirs(output_dir, exist_ok=True)

    # Use sample for plotting if dataset is too large
    sample_size = min(15000, len(df))
    if len(df) > sample_size:
        plot_df = df.sample(n=sample_size, random_state=42)
        print(f" Using sample of {sample_size:,} records for plotting")
    else:
        plot_df = df

    plt.style.use('default')

    # Timestamp for unique filenames
    timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')

    # 1. Top fuel types by total generation
    if 'fueltype' in plot_df.columns:
        plt.figure(figsize=(10, 6))
        fuel_totals = plot_df.groupby('fueltype')['value'].sum().sort_values(ascending=False).head(10)
        fuel_totals.plot(kind='bar', color='skyblue', edgecolor='black')
        plt.title('Top 10 Fuel Types by Total Generation', fontweight='bold', fontsize=14)
        plt.ylabel('Total Generation (MWh)')
        plt.xlabel('Fuel Type')
        plt.xticks(rotation=45)
        plt.tight_layout()
        if save_plots:
            filename = f"{output_dir}/top_fuel_types_{timestamp}.png"
            plt.savefig(filename, dpi=300, bbox_inches='tight')
            print(f" Saved: {filename}")
        plt.show()
        plt.close()

    # 2. Fuel category pie chart
    if 'fuel_category' in plot_df.columns:
        plt.figure(figsize=(8, 8))
        category_totals = plot_df.groupby('fuel_category')['value'].sum()
        colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
        plt.pie(category_totals.values, labels=category_totals.index, autopct='%1.1f%%',
                colors=colors, startangle=90)
        plt.title('Generation by Fuel Category', fontweight='bold', fontsize=14)
        plt.tight_layout()
        if save_plots:
            filename = f"{output_dir}/fuel_category_pie_{timestamp}.png"
            plt.savefig(filename, dpi=300, bbox_inches='tight')
            print(f" Saved: {filename}")
        plt.show()
        plt.close()

    # 3. Time series of top 5 fuel types
    if 'fueltype' in plot_df.columns:
        plt.figure(figsize=(12, 6))
        top_5_fuels = plot_df.groupby('fueltype')['value'].sum().nlargest(5).index
        for i, fuel in enumerate(top_5_fuels):
            fuel_data = plot_df[plot_df['fueltype'] == fuel]
            if len(fuel_data) > 200:
                fuel_data = fuel_data.sample(n=200, random_state=42).sort_values('period')
            plt.plot(fuel_data['period'], fuel_data['value'], label=fuel, alpha=0.8, linewidth=2)
        plt.title('Generation Time Series - Top 5 Fuel Types (Sample)', fontweight='bold', fontsize=14)
        plt.xlabel('Date')
        plt.ylabel('Daily Generation (MWh)')
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.xticks(rotation=45)
        plt.tight_layout()
        if save_plots:
            filename = f"{output_dir}/top_5_fuel_timeseries_{timestamp}.png"
            plt.savefig(filename, dpi=300, bbox_inches='tight')
            print(f" Saved: {filename}")
        plt.show()
        plt.close()

    # 4. Seasonal patterns by fuel category
    if 'season' in plot_df.columns and 'fuel_category' in plot_df.columns:
        plt.figure(figsize=(10, 6))
        seasonal_category = plot_df.groupby(['season', 'fuel_category'])['value'].mean().unstack(fill_value=0)
        seasonal_category.plot(kind='bar', stacked=True)
        plt.title('Seasonal Generation by Fuel Category', fontweight='bold', fontsize=14)
        plt.ylabel('Average Generation (MWh)')
        plt.xlabel('Season')
        plt.xticks(rotation=45)
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.tight_layout()
        if save_plots:
            filename = f"{output_dir}/seasonal_fuel_category_{timestamp}.png"
            plt.savefig(filename, dpi=300, bbox_inches='tight')
            print(f" Saved: {filename}")
        plt.show()
        plt.close()

    # 5. Monthly patterns for top fuel types
    if 'month' in plot_df.columns and 'fueltype' in plot_df.columns:
        plt.figure(figsize=(10, 6))
        top_3_fuels = plot_df.groupby('fueltype')['value'].sum().nlargest(3).index
        monthly_fuel = plot_df[plot_df['fueltype'].isin(top_3_fuels)].groupby(['month', 'fueltype'])['value'].mean().unstack(fill_value=0)
        monthly_fuel.plot(kind='line', marker='o')
        plt.title('Monthly Generation - Top 3 Fuel Types', fontweight='bold', fontsize=14)
        plt.xlabel('Month')
        plt.ylabel('Average Generation (MWh)')
        plt.xticks(range(1, 13))
        plt.legend()
        plt.tight_layout()
        if save_plots:
            filename = f"{output_dir}/monthly_top_3_fuel_{timestamp}.png"
            plt.savefig(filename, dpi=300, bbox_inches='tight')
            print(f" Saved: {filename}")
        plt.show()
        plt.close()

    # 6. Day of week patterns by fuel category
    if 'day_of_week' in plot_df.columns and 'fuel_category' in plot_df.columns:
        plt.figure(figsize=(10, 6))
        dow_category = plot_df.groupby(['day_of_week', 'fuel_category'])['value'].mean().unstack(fill_value=0)
        dow_category.plot(kind='bar', stacked=True)
        plt.title('Day of Week Generation by Fuel Category', fontweight='bold', fontsize=14)
        plt.xlabel('Day of Week')
        plt.ylabel('Average Generation (MWh)')
        day_names = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
        plt.xticks(range(len(day_names)), day_names, rotation=45)
        plt.legend()
        plt.tight_layout()
        if save_plots:
            filename = f"{output_dir}/day_of_week_fuel_category_{timestamp}.png"
            plt.savefig(filename, dpi=300, bbox_inches='tight')
            print(f" Saved: {filename}")
        plt.show()
        plt.close()

    # 7. Generation distribution by fuel category
    if 'fuel_category' in plot_df.columns:
        plt.figure(figsize=(10, 6))
        categories = plot_df['fuel_category'].unique()
        colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
        for i, category in enumerate(categories):
            category_data = plot_df[plot_df['fuel_category'] == category]['value']
            plt.hist(category_data, bins=30, alpha=0.7, label=category,
                    color=colors[i % len(colors)], density=True)
        plt.title('Generation Distribution by Fuel Category', fontweight='bold', fontsize=14)
        plt.xlabel('Generation (MWh)')
        plt.ylabel('Density')
        plt.legend()
        plt.yscale('log')
        plt.tight_layout()
        if save_plots:
            filename = f"{output_dir}/generation_distribution_fuel_category_{timestamp}.png"
            plt.savefig(filename, dpi=300, bbox_inches='tight')
            print(f" Saved: {filename}")
        plt.show()
        plt.close()

    # 8. Weekend vs weekday comparison by fuel category
    if 'is_weekend' in plot_df.columns and 'fuel_category' in plot_df.columns:
        plt.figure(figsize=(10, 6))
        weekend_comparison = plot_df.groupby(['fuel_category', 'is_weekend'])['value'].mean().unstack(fill_value=0)
        weekend_comparison.columns = ['Weekday', 'Weekend']
        weekend_comparison.plot(kind='bar')
        plt.title('Weekend vs Weekday Generation', fontweight='bold', fontsize=14)
        plt.ylabel('Average Generation (MWh)')
        plt.xlabel('Fuel Category')
        plt.xticks(rotation=45)
        plt.legend()
        plt.tight_layout()
        if save_plots:
            filename = f"{output_dir}/weekend_vs_weekday_{timestamp}.png"
            plt.savefig(filename, dpi=300, bbox_inches='tight')
            print(f" Saved: {filename}")
        plt.show()
        plt.close()

    # 9. Fuel mix percentage over time (stacked area chart)
    if 'fuel_mix_percentage' in plot_df.columns and 'fueltype' in plot_df.columns:
        plt.figure(figsize=(12, 6))
        plot_df['year_month'] = plot_df['period'].dt.to_period('M')
        monthly_mix = plot_df.groupby(['year_month', 'fueltype'])['fuel_mix_percentage'].mean().unstack(fill_value=0)
        top_fuels = monthly_mix.mean().nlargest(6).index
        monthly_mix_top = monthly_mix[top_fuels]
        monthly_mix_top.index = monthly_mix_top.index.to_timestamp()
        plt.stackplot(monthly_mix_top.index, *[monthly_mix_top[col] for col in monthly_mix_top.columns],
                      labels=monthly_mix_top.columns, alpha=0.8)
        plt.title('Monthly Fuel Mix Evolution (Top 6 Fuel Types)', fontweight='bold', fontsize=14)
        plt.xlabel('Date')
        plt.ylabel('Fuel Mix Percentage (%)')
        plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.xticks(rotation=45)
        plt.tight_layout()
        if save_plots:
            filename = f"{output_dir}/monthly_fuel_mix_{timestamp}.png"
            plt.savefig(filename, dpi=300, bbox_inches='tight')
            print(f" Saved: {filename}")
        plt.show()
        plt.close()

    # 10. Regional fuel mix heatmap
    if 'respondent' in plot_df.columns and 'fueltype' in plot_df.columns:
        plt.figure(figsize=(10, 8))
        regional_fuel_mix = plot_df.groupby(['respondent', 'fueltype'])['value'].sum().unstack(fill_value=0)
        regional_fuel_mix_pct = regional_fuel_mix.div(regional_fuel_mix.sum(axis=1), axis=0) * 100
        top_regions = plot_df.groupby('respondent')['value'].sum().nlargest(8).index
        top_fuels = plot_df.groupby('fueltype')['value'].sum().nlargest(6).index
        heatmap_data = regional_fuel_mix_pct.loc[top_regions, top_fuels]
        sns.heatmap(heatmap_data, annot=True, fmt='.1f', cmap='YlOrRd', cbar_kws={'label': 'Percentage (%)'})
        plt.title('Regional Fuel Mix Heatmap (%)', fontweight='bold', fontsize=14)
        plt.xlabel('Fuel Type')
        plt.ylabel('Region')
        plt.tight_layout()
        if save_plots:
            filename = f"{output_dir}/regional_fuel_mix_heatmap_{timestamp}.png"
            plt.savefig(filename, dpi=300, bbox_inches='tight')
            print(f" Saved: {filename}")
        plt.show()
        plt.close()

    # 11. Renewable vs non-renewable trends
    if 'is_renewable' in plot_df.columns and 'period' in plot_df.columns:
        plt.figure(figsize=(12, 6))
        plot_df['year_month'] = plot_df['period'].dt.to_period('M')
        renewable_trends = plot_df.groupby(['year_month', 'is_renewable'])['value'].sum().unstack(fill_value=0)
        expected_cols = [0, 1]
        for col in expected_cols:
            if col not in renewable_trends.columns:
                renewable_trends[col] = 0
        renewable_trends = renewable_trends[expected_cols]
        renewable_trends.columns = ['Non-Renewable', 'Renewable']
        renewable_trends.index = renewable_trends.index.to_timestamp()
        renewable_trends.plot(kind='area', stacked=True, alpha=0.7, color=['#ff7f7f', '#90EE90'])
        plt.title('Monthly Renewable vs Non-Renewable Generation Trends', fontweight='bold', fontsize=14)
        plt.xlabel('Date')
        plt.ylabel('Total Generation (MWh)')
        plt.legend()
        plt.xticks(rotation=45)
        plt.tight_layout()
        if save_plots:
            filename = f"{output_dir}/renewable_vs_non_renewable_{timestamp}.png"
            plt.savefig(filename, dpi=300, bbox_inches='tight')
            print(f" Saved: {filename}")
        plt.show()
        plt.close()

    # 12. Year-over-year comparison by fuel category
    if 'year' in plot_df.columns and 'fuel_category' in plot_df.columns:
        plt.figure(figsize=(10, 6))
        yearly_category = plot_df.groupby(['year', 'fuel_category'])['value'].sum().unstack(fill_value=0)
        yearly_category.plot(kind='bar', stacked=True)
        plt.title('Yearly Generation by Fuel Category', fontweight='bold', fontsize=14)
        plt.xlabel('Year')
        plt.ylabel('Total Generation (MWh)')
        plt.xticks(rotation=0)
        plt.legend()
        plt.tight_layout()
        if save_plots:
            filename = f"{output_dir}/yearly_fuel_category_{timestamp}.png"
            plt.savefig(filename, dpi=300, bbox_inches='tight')
            print(f" Saved: {filename}")
        plt.show()
        plt.close()

    # 13. Correlation heatmap of top fuel types
    if 'fueltype' in plot_df.columns and 'period' in plot_df.columns:
        plt.figure(figsize=(8, 8))
        top_5_fuels = plot_df.groupby('fueltype')['value'].sum().nlargest(5).index
        fuel_pivot = plot_df[plot_df['fueltype'].isin(top_5_fuels)].pivot_table(
            index='period', columns='fueltype', values='value', fill_value=0
        )
        correlation_matrix = fuel_pivot.corr()
        sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm',
                    center=0, cbar_kws={'label': 'Correlation'})
        plt.title('Fuel Type Correlation Matrix', fontweight='bold', fontsize=14)
        plt.tight_layout()
        if save_plots:
            filename = f"{output_dir}/fuel_type_correlation_{timestamp}.png"
            plt.savefig(filename, dpi=300, bbox_inches='tight')
            print(f" Saved: {filename}")
        plt.show()
        plt.close()

    # 14. Seasonal renewable percentage
    if 'season' in plot_df.columns and 'is_renewable' in plot_df.columns:
        plt.figure(figsize=(10, 6))
        seasonal_renewable = plot_df.groupby(['season', 'is_renewable'])['value'].sum().unstack(fill_value=0)
        seasonal_renewable_pct = seasonal_renewable.div(seasonal_renewable.sum(axis=1), axis=0) * 100
        if 1 in seasonal_renewable_pct.columns:
            seasonal_renewable_pct[1].plot(kind='bar', color='lightgreen', edgecolor='black')
            plt.title('Seasonal Renewable Energy Percentage', fontweight='bold', fontsize=14)
            plt.ylabel('Renewable Percentage (%)')
            plt.xlabel('Season')
            plt.xticks(rotation=45)
        else:
            plt.text(0.5, 0.5, 'No Renewable Data in Sample', horizontalalignment='center',
                     verticalalignment='center', transform=plt.gca().transAxes)
            plt.title('Seasonal Renewable Energy Percentage', fontweight='bold', fontsize=14)
        plt.tight_layout()
        if save_plots:
            filename = f"{output_dir}/seasonal_renewable_percentage_{timestamp}.png"
            plt.savefig(filename, dpi=300, bbox_inches='tight')
            print(f" Saved: {filename}")
        plt.show()
        plt.close()

    # 15. Fuel type volatility comparison
    if 'fueltype' in plot_df.columns:
        plt.figure(figsize=(10, 6))
        fuel_volatility = plot_df.groupby('fueltype')['value'].std().sort_values(ascending=False).head(10)
        fuel_volatility.plot(kind='bar', color='orange', edgecolor='black')
        plt.title('Fuel Type Generation Volatility (Std Dev)', fontweight='bold', fontsize=14)
        plt.ylabel('Standard Deviation (MWh)')
        plt.xlabel('Fuel Type')
        plt.xticks(rotation=45)
        plt.tight_layout()
        if save_plots:
            filename = f"{output_dir}/fuel_type_volatility_{timestamp}.png"
            plt.savefig(filename, dpi=300, bbox_inches='tight')
            print(f" Saved: {filename}")
        plt.show()
        plt.close()

    # 16. Holiday vs non-holiday generation patterns
    if 'is_holiday' in plot_df.columns and 'fuel_category' in plot_df.columns:
        plt.figure(figsize=(10, 6))
        holiday_comparison = plot_df.groupby(['fuel_category', 'is_holiday'])['value'].mean().unstack(fill_value=0)
        holiday_comparison.columns = ['Non-Holiday', 'Holiday']
        holiday_comparison.plot(kind='bar')
        plt.title('Holiday vs Non-Holiday Generation', fontweight='bold', fontsize=14)
        plt.ylabel('Average Generation (MWh)')
        plt.xlabel('Fuel Category')
        plt.xticks(rotation=45)
        plt.legend()
        plt.tight_layout()
        if save_plots:
            filename = f"{output_dir}/holiday_vs_non_holiday_{timestamp}.png"
            plt.savefig(filename, dpi=300, bbox_inches='tight')
            print(f" Saved: {filename}")
        plt.show()
        plt.close()

    # 17. Top regions by renewable percentage
    if 'respondent' in plot_df.columns and 'is_renewable' in plot_df.columns:
        plt.figure(figsize=(10, 6))
        regional_renewable = plot_df.groupby(['respondent', 'is_renewable'])['value'].sum().unstack(fill_value=0)
        regional_renewable_pct = regional_renewable.div(regional_renewable.sum(axis=1), axis=0) * 100
        if 1 in regional_renewable_pct.columns:
            top_renewable_regions = regional_renewable_pct[1].nlargest(10)
            top_renewable_regions.plot(kind='bar', color='lightblue', edgecolor='black')
            plt.title('Top 10 Regions by Renewable %', fontweight='bold', fontsize=14)
            plt.ylabel('Renewable Percentage (%)')
            plt.xlabel('Region')
            plt.xticks(rotation=45)
        else:
            plt.text(0.5, 0.5, 'No Renewable Data in Sample', horizontalalignment='center',
                     verticalalignment='center', transform=plt.gca().transAxes)
            plt.title('Top 10 Regions by Renewable %', fontweight='bold', fontsize=14)
        plt.tight_layout()
        if save_plots:
            filename = f"{output_dir}/top_regions_renewable_percentage_{timestamp}.png"
            plt.savefig(filename, dpi=300, bbox_inches='tight')
            print(f" Saved: {filename}")
        plt.show()
        plt.close()

    print("\n All individual visualizations created and saved!")

# Example usage
def main_visualization():
    # Load the CSV file
    csv_file = "/content/drive/MyDrive/DS200/Dataset/daily_fuel_type_data_2020-06-01_to_2022-06-01.csv"
    print(f" Loading data from {csv_file}...")
    try:
        df = pd.read_csv(csv_file)
        # Ensure 'period' is datetime
        if 'period' in df.columns:
            df['period'] = pd.to_datetime(df['period'])
        print(f" Data loaded: {df.shape[0]:,} rows, {df.shape[1]} columns")
    except FileNotFoundError:
        print(f" File {csv_file} not found!")
        return
    except Exception as e:
        print(f" Error loading CSV: {e}")
        return

    # Create visualizations
    create_individual_fuel_type_visualizations(df, save_plots=True, output_dir="fuel_type_plots")

if __name__ == "__main__":
    main_visualization()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import os

def load_csv_data(file_path: str) -> pd.DataFrame:
    """Load data from CSV file"""
    print(f" Loading data from {file_path}...")
    try:
        df = pd.read_csv(file_path)
        # Ensure 'period' is datetime
        if 'period' in df.columns:
            df['period'] = pd.to_datetime(df['period'])
        print(f" Loaded {len(df):,} records successfully")
        return df
    except Exception as e:
        print(f" Error loading CSV: {e}")
        return pd.DataFrame()

def setup_plot_style():
    """Set up consistent plot style"""
    plt.style.use('default')
    return {
        'figsize': (12, 6),
        'title_fontsize': 14,
        'label_fontsize': 12,
        'tick_fontsize': 10,
        'dpi': 300
    }

def save_plot(fig, filename: str, plot_config: dict):
    """Save plot with consistent formatting and display it"""
    output_dir = 'visualizations'
    os.makedirs(output_dir, exist_ok=True)
    full_path = os.path.join(output_dir, filename)
    fig.savefig(full_path, dpi=plot_config['dpi'], bbox_inches='tight')
    print(f" Plot saved to {full_path}")
    plt.show()  # Display the plot
    plt.close(fig)

def plot_regional_comparison(df: pd.DataFrame, plot_config: dict):
    """Plot top 10 regions by total demand"""
    if 'respondent' not in df.columns:
        print(" No 'respondent' column found")
        return

    fig, ax = plt.subplots(figsize=plot_config['figsize'])
    regional_totals = df.groupby('respondent')['value'].sum().sort_values(ascending=False).head(10)
    regional_totals.plot(kind='bar', ax=ax, color='skyblue', edgecolor='black')

    ax.set_title('Top Regions by Total Demand', fontsize=plot_config['title_fontsize'], fontweight='bold')
    ax.set_ylabel('Total Demand (MWh)', fontsize=plot_config['label_fontsize'])
    ax.set_xlabel('Region', fontsize=plot_config['label_fontsize'])
    ax.tick_params(axis='x', rotation=45, labelsize=plot_config['tick_fontsize'])

    save_plot(fig, f'regional_comparison_{datetime.now().strftime("%Y%m%d_%H%M%S")}.png', plot_config)

def plot_time_series(df: pd.DataFrame, plot_config: dict):
    """Plot time series for top 5 regions"""
    if 'respondent' not in df.columns or 'period' not in df.columns:
        print(" Required columns missing")
        return

    fig, ax = plt.subplots(figsize=plot_config['figsize'])
    top_5_regions = df.groupby('respondent')['value'].sum().nlargest(5).index

    for region in top_5_regions:
        region_data = df[df['respondent'] == region]
        if len(region_data) > 100:
            region_data = region_data.sample(n=100, random_state=42).sort_values('period')
        ax.plot(region_data['period'], region_data['value'], label=region, alpha=0.7, linewidth=2)

    ax.set_title('Demand Time Series - Top 5 Regions', fontsize=plot_config['title_fontsize'], fontweight='bold')
    ax.set_xlabel('Date', fontsize=plot_config['label_fontsize'])
    ax.set_ylabel('Daily Demand (MWh)', fontsize=plot_config['label_fontsize'])
    ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    ax.tick_params(axis='x', rotation=45, labelsize=plot_config['tick_fontsize'])

    save_plot(fig, f'time_series_{datetime.now().strftime("%Y%m%d_%H%M%S")}.png', plot_config)

def plot_seasonal_patterns(df: pd.DataFrame, plot_config: dict):
    """Plot average demand by season"""
    if 'season' not in df.columns:
        print(" No 'season' column found")
        return

    fig, ax = plt.subplots(figsize=plot_config['figsize'])
    seasonal_avg = df.groupby('season')['value'].mean()
    seasonal_avg.plot(kind='bar', ax=ax, color=['lightblue', 'lightgreen', 'orange', 'lightcoral'])

    ax.set_title('Average Demand by Season', fontsize=plot_config['title_fontsize'], fontweight='bold')
    ax.set_ylabel('Average Demand (MWh)', fontsize=plot_config['label_fontsize'])
    ax.set_xlabel('Season', fontsize=plot_config['label_fontsize'])
    ax.tick_params(axis='x', rotation=0, labelsize=plot_config['tick_fontsize'])

    save_plot(fig, f'seasonal_patterns_{datetime.now().strftime("%Y%m%d_%H%M%S")}.png', plot_config)

def plot_monthly_patterns(df: pd.DataFrame, plot_config: dict):
    """Plot average demand by month"""
    if 'month' not in df.columns:
        print(" No 'month' column found")
        return

    fig, ax = plt.subplots(figsize=plot_config['figsize'])
    monthly_avg = df.groupby('month')['value'].mean()
    month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

    ax.bar(range(1, 13), [monthly_avg.get(i, 0) for i in range(1, 13)],
           color='lightsteelblue', edgecolor='black')

    ax.set_title('Average Demand by Month', fontsize=plot_config['title_fontsize'], fontweight='bold')
    ax.set_xlabel('Month', fontsize=plot_config['label_fontsize'])
    ax.set_ylabel('Average Demand (MWh)', fontsize=plot_config['label_fontsize'])
    ax.set_xticks(range(1, 13))
    ax.set_xticklabels(month_names, rotation=45)
    ax.tick_params(axis='both', labelsize=plot_config['tick_fontsize'])

    save_plot(fig, f'monthly_patterns_{datetime.now().strftime("%Y%m%d_%H%M%S")}.png', plot_config)

def plot_day_of_week_patterns(df: pd.DataFrame, plot_config: dict):
    """Plot average demand by day of week"""
    if 'day_of_week' not in df.columns:
        print(" No 'day_of_week' column found")
        return

    fig, ax = plt.subplots(figsize=plot_config['figsize'])
    dow_avg = df.groupby('day_of_week')['value'].mean()
    day_names = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

    ax.bar(range(7), [dow_avg.get(i, 0) for i in range(7)],
           color='lightcoral', edgecolor='black')

    ax.set_title('Average Demand by Day of Week', fontsize=plot_config['title_fontsize'], fontweight='bold')
    ax.set_xlabel('Day of Week', fontsize=plot_config['label_fontsize'])
    ax.set_ylabel('Average Demand (MWh)', fontsize=plot_config['label_fontsize'])
    ax.set_xticks(range(7))
    ax.set_xticklabels(day_names)
    ax.tick_params(axis='both', labelsize=plot_config['tick_fontsize'])

    save_plot(fig, f'day_of_week_patterns_{datetime.now().strftime("%Y%m%d_%H%M%S")}.png', plot_config)

def plot_demand_distribution(df: pd.DataFrame, plot_config: dict):
    """Plot demand distribution histogram"""
    fig, ax = plt.subplots(figsize=plot_config['figsize'])
    ax.hist(df['value'], bins=50, alpha=0.7, edgecolor='black', color='lightgreen')

    ax.set_title('Demand Distribution', fontsize=plot_config['title_fontsize'], fontweight='bold')
    ax.set_xlabel('Daily Demand (MWh)', fontsize=plot_config['label_fontsize'])
    ax.set_ylabel('Frequency', fontsize=plot_config['label_fontsize'])
    ax.set_yscale('log')
    ax.tick_params(axis='both', labelsize=plot_config['tick_fontsize'])

    save_plot(fig, f'demand_distribution_{datetime.now().strftime("%Y%m%d_%H%M%S")}.png', plot_config)

def plot_weekend_vs_weekday(df: pd.DataFrame, plot_config: dict):
    """Plot boxplot comparing weekend vs weekday demand"""
    if 'is_weekend' not in df.columns:
        print(" No 'is_weekend' column found")
        return

    fig, ax = plt.subplots(figsize=plot_config['figsize'])
    weekend_data = df[df['is_weekend'] == 1]['value']
    weekday_data = df[df['is_weekend'] == 0]['value']

    boxes = ax.boxplot([weekday_data, weekend_data], labels=['Weekday', 'Weekend'], patch_artist=True)

    for box in boxes['boxes']:
        box.set_facecolor('lightblue' if boxes['boxes'].index(box) == 0 else 'lightcoral')

    ax.set_title('Weekend vs Weekday Demand', fontsize=plot_config['title_fontsize'], fontweight='bold')
    ax.set_ylabel('Daily Demand (MWh)', fontsize=plot_config['label_fontsize'])
    ax.tick_params(axis='both', labelsize=plot_config['tick_fontsize'])

    save_plot(fig, f'weekend_vs_weekday_{datetime.now().strftime("%Y%m%d_%H%M%S")}.png', plot_config)

def plot_yearly_comparison(df: pd.DataFrame, plot_config: dict):
    """Plot average demand by year"""
    if 'year' not in df.columns:
        print(" No 'year' column found")
        return

    fig, ax = plt.subplots(figsize=plot_config['figsize'])
    yearly_avg = df.groupby('year')['value'].mean()
    yearly_avg.plot(kind='bar', ax=ax, color='gold', edgecolor='black')

    ax.set_title('Average Demand by Year', fontsize=plot_config['title_fontsize'], fontweight='bold')
    ax.set_xlabel('Year', fontsize=plot_config['label_fontsize'])
    ax.set_ylabel('Average Demand (MWh)', fontsize=plot_config['label_fontsize'])
    ax.tick_params(axis='both', rotation=0, labelsize=plot_config['tick_fontsize'])

    save_plot(fig, f'yearly_comparison_{datetime.now().strftime("%Y%m%d_%H%M%S")}.png', plot_config)

def plot_regional_heatmap(df: pd.DataFrame, plot_config: dict):
    """Plot monthly demand heatmap by region"""
    if 'respondent' not in df.columns or 'month' not in df.columns:
        print(" Required columns missing")
        return

    fig, ax = plt.subplots(figsize=(plot_config['figsize'][0], plot_config['figsize'][1] + 2))
    pivot_data = df.groupby(['respondent', 'month'])['value'].mean().unstack(fill_value=0)

    top_regions = df.groupby('respondent')['value'].sum().nlargest(10).index
    pivot_data = pivot_data.loc[top_regions]

    sns.heatmap(pivot_data, annot=True, fmt='.0f', cmap='YlOrRd', ax=ax)

    ax.set_title('Monthly Demand Heatmap by Region (MWh)', fontsize=plot_config['title_fontsize'], fontweight='bold')
    ax.set_xlabel('Month', fontsize=plot_config['label_fontsize'])
    ax.set_ylabel('Region', fontsize=plot_config['label_fontsize'])
    month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    ax.set_xticklabels(month_labels)
    ax.tick_params(axis='both', labelsize=plot_config['tick_fontsize'])

    save_plot(fig, f'regional_heatmap_{datetime.now().strftime("%Y%m%d_%H%M%S")}.png', plot_config)

def create_separate_visualizations(df: pd.DataFrame, save_plots: bool = True):
    """Create separate visualizations for daily regional data"""
    print("\n Creating separate visualizations...")

    if df.empty:
        print(" Empty DataFrame. No visualizations created.")
        return

    # Sample data if too large
    sample_size = min(10000, len(df))
    if len(df) > sample_size:
        plot_df = df.sample(n=sample_size, random_state=42)
        print(f" Using sample of {sample_size:,} records for plotting")
    else:
        plot_df = df

    plot_config = setup_plot_style()

    # Create each visualization
    plot_regional_comparison(plot_df, plot_config)
    plot_time_series(plot_df, plot_config)
    plot_seasonal_patterns(plot_df, plot_config)
    plot_monthly_patterns(plot_df, plot_config)
    plot_day_of_week_patterns(plot_df, plot_config)
    plot_demand_distribution(plot_df, plot_config)
    plot_weekend_vs_weekday(plot_df, plot_config)
    plot_yearly_comparison(plot_df, plot_config)
    plot_regional_heatmap(plot_df, plot_config)

    print("\n All visualizations completed!")

def main():
    """Main function to load data and create visualizations"""
    file_path = "/content/drive/MyDrive/DS200/Dataset/daily_region_data_2020-06-01_to_2022-06-01.csv"
    df = load_csv_data(file_path)

    if not df.empty:
        create_separate_visualizations(df, save_plots=True)
    else:
        print(" No data loaded. Visualization aborted.")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8-whitegrid')

# Use a font that supports Vietnamese characters
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['axes.unicode_minus'] = False


# ---------- 1. EDA: D liu nhu cu in ---------- #
def eda_energy_demand(df):
    print("\n EDA - D liu Nhu cu in")
    cols = ['period', 'respondent', 'respondent-name', 'type', 'type-name',
            'timezone', 'timezone-description', 'value', 'value-units']
    df_energy = df[cols].copy()

    print(df_energy.info())
    print(df_energy.describe(include='all'))
    print(" Missing values:\n", df_energy.isna().sum())

    # Phn phi value
    plt.figure(figsize=(10,6))
    sns.histplot(df_energy['value'], bins=50, kde=True)
    plt.title('Phn phi nhu cu in')
    plt.xlabel('MWh')
    plt.ylabel('Tn sut')
    plt.show()

    # Top 10 vng
    top_regions = df_energy.groupby('respondent')['value'].sum().nlargest(10)
    top_regions.plot(kind='bar', figsize=(10,6), title='Top 10 vng c tng nhu cu in cao nht')
    plt.ylabel('Tng nhu cu (MWh)')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Xu hng thi gian
    df_energy['period'] = pd.to_datetime(df_energy['period'])
    daily_demand = df_energy.groupby('period')['value'].sum()
    daily_demand.plot(figsize=(15,6), title='Xu hng tng nhu cu in theo thi gian')
    plt.xlabel('Ngy')
    plt.ylabel('Tng nhu cu (MWh)')
    plt.grid(True)
    plt.show()

# ---------- 2. EDA: D liu Thi tit ---------- #
def eda_weather(df):
    print("\n EDA - D liu Thi tit")
    cols = ['time', 'region', 'tavg', 'tmin', 'tmax', 'prcp', 'wspd', 'pres', 'tsun']
    df_weather = df[cols].copy()

    print(df_weather.describe())
    print(" Missing values:\n", df_weather.isna().sum())

    # Phn phi cc bin thi tit
    weather_features = ['tavg', 'tmin', 'tmax', 'prcp', 'wspd', 'pres', 'tsun']
    df_weather[weather_features].hist(bins=30, figsize=(15,10))
    plt.suptitle('Phn phi cc c trng thi tit')
    plt.show()

    # Xu hng nhit  trung bnh
    df_weather['time'] = pd.to_datetime(df_weather['time'])
    avg_temp_time = df_weather.groupby('time')['tavg'].mean()
    avg_temp_time.plot(figsize=(15,6), title='Xu hng nhit  trung bnh theo thi gian')
    plt.ylabel('Nhit  trung bnh (C)')
    plt.grid(True)
    plt.show()

    # Heatmap
    corr_matrix = df_weather[weather_features].corr()
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
    plt.title('Tng quan cc c trng thi tit')
    plt.show()

# ---------- 3. EDA: D liu Nhin liu theo ngy ---------- #
def eda_fuel_type(df):
    print("\n EDA - D liu Nhin liu theo ngy")
    print(df.info())
    print(" Missing values:\n", df.isna().sum())

    # Tng sn lng theo loi nhin liu
    fuel_summary = df.groupby('fueltype')['value'].sum().sort_values(ascending=False)
    fuel_summary.plot(kind='bar', figsize=(12,6), title='Tng sn lng theo loi nhin liu')
    plt.ylabel('Sn lng (MWh)')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Xu hng theo thi gian cho top 5 loi nhin liu
    df['period'] = pd.to_datetime(df['period'])
    main_fuels = df['fueltype'].value_counts().head(5).index

    plt.figure(figsize=(15,8))
    for fuel in main_fuels:
        subset = df[df['fueltype'] == fuel]
        fuel_trend = subset.groupby('period')['value'].sum()
        plt.plot(fuel_trend.index, fuel_trend.values, label=fuel)

    plt.title('Xu hng sn lng theo thi gian ca top 5 loi nhin liu')
    plt.xlabel('Ngy')
    plt.ylabel('Sn lng (MWh)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# ---------- Hm chnh ---------- #
def main():
    # ng dn file
    path_energy_weather = '/content/drive/MyDrive/DS200/Dataset/energy_weather_data_2020-06-01_to_2022-06-01_US_Center.csv'
    path_fuel_type = '/content/drive/MyDrive/DS200/Dataset/daily_fuel_type_data_2020-06-01_to_2022-06-01.csv'

    # c d liu
    print(" ang c d liu...")
    df_energy_weather = pd.read_csv(path_energy_weather)
    df_fuel_type = pd.read_csv(path_fuel_type)

    # Thc hin EDA
    eda_energy_demand(df_energy_weather)
    eda_weather(df_energy_weather)
    eda_fuel_type(df_fuel_type)

if __name__ == '__main__':
    main()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Set style for better visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

class OriginalDataEDA:
    def __init__(self):
        self.energy_weather_data = None
        self.daily_fuel_data = None
        self.removed_columns = {'energy_weather': [], 'daily_fuel': []}

        # nh ngha cc ct gc cn gi li
        self.original_energy_weather_cols = [
            # Phn EIA (nng lng)
            'period', 'respondent', 'respondent-name', 'type', 'type-name',
            'timezone', 'timezone-description', 'value', 'value-units',
            # Phn Meteostat (thi tit)
            'time', 'region', 'tavg', 'tmin', 'tmax', 'prcp', 'wspd', 'pres', 'tsun'
        ]

        self.original_daily_fuel_cols = [
            'period', 'respondent', 'respondent-name', 'fueltype', 'type-name', 'value', 'units'
        ]

    def filter_original_columns(self, df, original_cols, dataset_name):
        """Lc v ch gi li cc ct d liu gc"""
        print(f"\n FILTERING {dataset_name.upper()} TO ORIGINAL COLUMNS ONLY")
        print("-" * 50)

        # Cc ct hin c trong dataset
        current_cols = list(df.columns)
        print(f"Tng s ct hin ti: {len(current_cols)}")

        # Cc ct gc tn ti trong dataset
        existing_original_cols = [col for col in original_cols if col in current_cols]
        print(f"S ct gc tn ti: {len(existing_original_cols)}")

        # Cc ct s b loi b (feature engineered columns)
        removed_cols = [col for col in current_cols if col not in original_cols]
        self.removed_columns[dataset_name] = removed_cols

        if removed_cols:
            print(f"  loi b {len(removed_cols)} ct feature engineered:")
            for i, col in enumerate(removed_cols, 1):
                print(f"   {i:2d}. {col}")
        else:
            print(" Khng c ct no b loi b - dataset ch cha d liu gc")

        # Cc ct gc b thiu
        missing_original_cols = [col for col in original_cols if col not in current_cols]
        if missing_original_cols:
            print(f"  Cnh bo: {len(missing_original_cols)} ct gc b thiu:")
            for col in missing_original_cols:
                print(f"   - {col}")

        # Tr v dataset  c lc
        filtered_df = df[existing_original_cols].copy()
        print(f" Dataset sau khi lc: {filtered_df.shape[0]} rows  {filtered_df.shape[1]} columns")

        return filtered_df

    def load_and_filter_data(self, energy_weather_path, daily_fuel_path):
        """Load v filter d liu ch gi li ct gc"""
        print("=" * 60)
        print(" LOADING AND FILTERING DATA TO ORIGINAL COLUMNS")
        print("=" * 60)

        # Load raw data
        print("Loading raw datasets...")
        raw_energy_weather = pd.read_csv(energy_weather_path)
        raw_daily_fuel = pd.read_csv(daily_fuel_path)
        print(" Raw data loaded successfully")

        # Filter to original columns only
        self.energy_weather_data = self.filter_original_columns(
            raw_energy_weather, self.original_energy_weather_cols, 'energy_weather'
        )

        self.daily_fuel_data = self.filter_original_columns(
            raw_daily_fuel, self.original_daily_fuel_cols, 'daily_fuel'
        )

        print("\n DATA FILTERING SUMMARY:")
        print("-" * 30)
        print(f"Energy Weather Data: {raw_energy_weather.shape}  {self.energy_weather_data.shape}")
        print(f"Daily Fuel Data: {raw_daily_fuel.shape}  {self.daily_fuel_data.shape}")

    def basic_info_analysis(self):
        """Phn tch thng tin c bn cho d liu gc"""
        print("\n" + "=" * 60)
        print("1. BASIC INFORMATION ANALYSIS (ORIGINAL DATA ONLY)")
        print("=" * 60)

        datasets = {
            'Energy Weather Data (Original)': self.energy_weather_data,
            'Daily Fuel Type Data (Original)': self.daily_fuel_data
        }

        for name, df in datasets.items():
            print(f"\n {name}:")
            print(f"   Shape: {df.shape}")
            print(f"   Memory usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

            # Phn loi ct theo loi d liu
            numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
            datetime_cols = []
            categorical_cols = []

            for col in df.columns:
                if col not in numeric_cols:
                    # Th convert sang datetime
                    try:
                        pd.to_datetime(df[col].dropna().iloc[:100])
                        datetime_cols.append(col)
                    except:
                        categorical_cols.append(col)

            print(f"    Numeric columns ({len(numeric_cols)}): {numeric_cols}")
            print(f"    DateTime columns ({len(datetime_cols)}): {datetime_cols}")
            print(f"     Categorical columns ({len(categorical_cols)}): {categorical_cols}")

            # Missing values
            missing = df.isnull().sum()
            if missing.sum() > 0:
                print(f"    Missing values:")
                for col, miss_count in missing[missing > 0].items():
                    miss_percent = (miss_count / len(df)) * 100
                    print(f"      {col}: {miss_count} ({miss_percent:.1f}%)")
            else:
                print(f"    No missing values")

            # Duplicates
            duplicates = df.duplicated().sum()
            print(f"    Duplicate rows: {duplicates}")

            print("-" * 50)

    def original_data_statistics(self):
        """Thng k m t cho d liu gc"""
        print("\n" + "=" * 60)
        print("2. DESCRIPTIVE STATISTICS (ORIGINAL DATA)")
        print("=" * 60)

        # Energy Weather Data
        print("\n Energy Weather Data - Numeric Variables:")
        numeric_cols_energy = self.energy_weather_data.select_dtypes(include=[np.number]).columns
        if len(numeric_cols_energy) > 0:
            stats = self.energy_weather_data[numeric_cols_energy].describe()
            print(stats.round(3))

            # Specific analysis for energy demand (value column)
            if 'value' in numeric_cols_energy:
                print(f"\n Energy Demand (value) Detailed Analysis:")
                energy_values = self.energy_weather_data['value'].dropna()
                print(f"   Total records: {len(energy_values):,}")
                print(f"   Mean: {energy_values.mean():,.2f}")
                print(f"   Median: {energy_values.median():,.2f}")
                print(f"   Mode: {energy_values.mode().iloc[0] if len(energy_values.mode()) > 0 else 'N/A'}")
                print(f"   Coefficient of Variation: {(energy_values.std()/energy_values.mean()*100):.2f}%")

                # Outlier detection
                Q1 = energy_values.quantile(0.25)
                Q3 = energy_values.quantile(0.75)
                IQR = Q3 - Q1
                outliers = energy_values[(energy_values < Q1 - 1.5*IQR) | (energy_values > Q3 + 1.5*IQR)]
                print(f"   Outliers (IQR method): {len(outliers)} ({len(outliers)/len(energy_values)*100:.2f}%)")

        # Weather variables analysis
        weather_cols = ['tavg', 'tmin', 'tmax', 'prcp', 'wspd', 'pres', 'tsun']
        available_weather = [col for col in weather_cols if col in self.energy_weather_data.columns]

        if available_weather:
            print(f"\n  Weather Variables Analysis:")
            weather_stats = self.energy_weather_data[available_weather].describe()
            print(weather_stats.round(3))

        # Daily Fuel Data
        print("\n Daily Fuel Type Data - Numeric Variables:")
        numeric_cols_fuel = self.daily_fuel_data.select_dtypes(include=[np.number]).columns
        if len(numeric_cols_fuel) > 0:
            fuel_stats = self.daily_fuel_data[numeric_cols_fuel].describe()
            print(fuel_stats.round(3))

    def original_time_analysis(self):
        """Phn tch thi gian cho d liu gc"""
        print("\n" + "=" * 60)
        print("3. TIME SERIES ANALYSIS (ORIGINAL DATA)")
        print("=" * 60)

        # Convert time columns
        time_conversions = []

        for col in ['period', 'time']:
            if col in self.energy_weather_data.columns:
                try:
                    self.energy_weather_data[col] = pd.to_datetime(self.energy_weather_data[col])
                    time_conversions.append(f"energy_weather.{col}")
                except:
                    print(f"  Khng th convert {col} trong energy_weather_data sang datetime")

        if 'period' in self.daily_fuel_data.columns:
            try:
                self.daily_fuel_data['period'] = pd.to_datetime(self.daily_fuel_data['period'])
                time_conversions.append("daily_fuel.period")
            except:
                print("  Khng th convert period trong daily_fuel_data sang datetime")

        if time_conversions:
            print(f" Converted to datetime: {', '.join(time_conversions)}")

        # Time range analysis
        print(f"\n Time Range Analysis:")

        # Energy Weather data time range
        if 'period' in self.energy_weather_data.columns:
            period_range = self.energy_weather_data['period'].dropna()
            if len(period_range) > 0:
                print(f"   Energy Data (period): {period_range.min()} to {period_range.max()}")
                print(f"   Duration: {period_range.max() - period_range.min()}")

        if 'time' in self.energy_weather_data.columns:
            time_range = self.energy_weather_data['time'].dropna()
            if len(time_range) > 0:
                print(f"   Weather Data (time): {time_range.min()} to {time_range.max()}")
                print(f"   Duration: {time_range.max() - time_range.min()}")

        # Daily Fuel data time range
        if 'period' in self.daily_fuel_data.columns:
            fuel_period = self.daily_fuel_data['period'].dropna()
            if len(fuel_period) > 0:
                print(f"   Fuel Data (period): {fuel_period.min()} to {fuel_period.max()}")
                print(f"   Duration: {fuel_period.max() - fuel_period.min()}")

        # Data frequency analysis
        print(f"\n Data Frequency Analysis:")
        datasets = [
            ('Energy Weather', self.energy_weather_data, 'period'),
            ('Daily Fuel', self.daily_fuel_data, 'period')
        ]

        for name, df, time_col in datasets:
            if time_col in df.columns:
                time_data = df[time_col].dropna()
                if len(time_data) > 1:
                    # Tnh time differences
                    time_diffs = time_data.sort_values().diff().dropna()
                    most_common_freq = time_diffs.mode()
                    if len(most_common_freq) > 0:
                        print(f"   {name}: Most common frequency = {most_common_freq.iloc[0]}")
                    print(f"   {name}: Total records = {len(time_data):,}")

    def original_categorical_analysis(self):
        """Phn tch cc bin categorical trong d liu gc"""
        print("\n" + "=" * 60)
        print("4. CATEGORICAL VARIABLES ANALYSIS (ORIGINAL DATA)")
        print("=" * 60)

        # Energy Weather categorical analysis
        print(f"\n  Energy Weather Data - Categorical Variables:")
        categorical_cols = self.energy_weather_data.select_dtypes(include=['object']).columns

        for col in categorical_cols:
            if col not in ['period', 'time']:  # Skip time columns
                unique_vals = self.energy_weather_data[col].nunique()
                print(f"\n   {col}:")
                print(f"     Unique values: {unique_vals}")

                # Show value counts for columns with reasonable number of unique values
                if unique_vals <= 20:
                    value_counts = self.energy_weather_data[col].value_counts()
                    for val, count in value_counts.head(10).items():
                        percentage = (count / len(self.energy_weather_data)) * 100
                        print(f"     {val}: {count} ({percentage:.1f}%)")
                else:
                    print(f"     Too many unique values to display (showing top 5)")
                    value_counts = self.energy_weather_data[col].value_counts()
                    for val, count in value_counts.head(5).items():
                        percentage = (count / len(self.energy_weather_data)) * 100
                        print(f"     {val}: {count} ({percentage:.1f}%)")

        # Daily Fuel categorical analysis
        print(f"\n Daily Fuel Data - Categorical Variables:")
        fuel_categorical = self.daily_fuel_data.select_dtypes(include=['object']).columns

        for col in fuel_categorical:
            if col != 'period':  # Skip time column
                unique_vals = self.daily_fuel_data[col].nunique()
                print(f"\n   {col}:")
                print(f"     Unique values: {unique_vals}")

                if unique_vals <= 20:
                    value_counts = self.daily_fuel_data[col].value_counts()
                    for val, count in value_counts.head(10).items():
                        percentage = (count / len(self.daily_fuel_data)) * 100
                        print(f"     {val}: {count} ({percentage:.1f}%)")

    def original_correlation_analysis(self):
        """Phn tch correlation cho d liu gc"""
        print("\n" + "=" * 60)
        print("5. CORRELATION ANALYSIS (ORIGINAL DATA ONLY)")
        print("=" * 60)

        # Weather-Energy correlations
        weather_cols = ['tavg', 'tmin', 'tmax', 'prcp', 'wspd', 'pres', 'tsun']
        available_weather = [col for col in weather_cols if col in self.energy_weather_data.columns]

        if 'value' in self.energy_weather_data.columns and available_weather:
            print(f"\n  Energy Demand vs Weather Correlations (Original Data):")

            energy_weather_data = self.energy_weather_data[['value'] + available_weather].dropna()

            if len(energy_weather_data) > 0:
                correlations = {}
                for weather_var in available_weather:
                    corr = energy_weather_data['value'].corr(energy_weather_data[weather_var])
                    correlations[weather_var] = corr

                # Sort by absolute correlation
                sorted_corr = sorted(correlations.items(), key=lambda x: abs(x[1]), reverse=True)

                print(f"   Based on {len(energy_weather_data):,} complete records:")
                for var, corr in sorted_corr:
                    strength = "Strong" if abs(corr) > 0.7 else "Moderate" if abs(corr) > 0.3 else "Weak"
                    direction = "Positive" if corr > 0 else "Negative"
                    print(f"     {var:>6}: {corr:>7.3f} ({strength} {direction})")

            # Weather variables inter-correlations
            if len(available_weather) > 1:
                print(f"\n  Weather Variables Inter-correlations:")
                weather_corr_matrix = energy_weather_data[available_weather].corr()

                # Show strongest correlations (excluding self-correlations)
                strong_corrs = []
                for i, var1 in enumerate(available_weather):
                    for j, var2 in enumerate(available_weather):
                        if i < j:  # Avoid duplicates and self-correlations
                            corr = weather_corr_matrix.loc[var1, var2]
                            if abs(corr) > 0.5:  # Only show moderate to strong correlations
                                strong_corrs.append((var1, var2, corr))

                strong_corrs.sort(key=lambda x: abs(x[2]), reverse=True)

                if strong_corrs:
                    print(f"   Strong inter-correlations (|r| > 0.5):")
                    for var1, var2, corr in strong_corrs:
                        print(f"     {var1}  {var2}: {corr:.3f}")
                else:
                    print(f"   No strong inter-correlations found (all |r|  0.5)")

    def create_original_data_visualizations(self):
        """To visualizations cho d liu gc"""
        print("\n" + "=" * 60)
        print("6. CREATING VISUALIZATIONS (ORIGINAL DATA)")
        print("=" * 60)

        plt.rcParams['figure.figsize'] = (15, 10)

        # 1. Energy demand analysis
        if 'value' in self.energy_weather_data.columns:
            fig, axes = plt.subplots(2, 2, figsize=(16, 12))
            fig.suptitle('Energy Demand Analysis (Original Data)', fontsize=16, fontweight='bold')

            # Time series (if period exists)
            if 'period' in self.energy_weather_data.columns:
                axes[0, 0].plot(self.energy_weather_data['period'], self.energy_weather_data['value'],
                               alpha=0.7, linewidth=1)
                axes[0, 0].set_title('Energy Demand Over Time')
                axes[0, 0].set_xlabel('Period')
                axes[0, 0].set_ylabel('Energy Demand')
                axes[0, 0].tick_params(axis='x', rotation=45)

            # Distribution
            energy_clean = self.energy_weather_data['value'].dropna()
            axes[0, 1].hist(energy_clean, bins=50, alpha=0.7, edgecolor='black')
            axes[0, 1].set_title('Energy Demand Distribution')
            axes[0, 1].set_xlabel('Energy Demand')
            axes[0, 1].set_ylabel('Frequency')

            # Box plot
            axes[1, 0].boxplot(energy_clean)
            axes[1, 0].set_title('Energy Demand Box Plot')
            axes[1, 0].set_ylabel('Energy Demand')

            # QQ plot (approximate)
            from scipy import stats
            stats.probplot(energy_clean, dist="norm", plot=axes[1, 1])
            axes[1, 1].set_title('Q-Q Plot (Normal Distribution)')

            plt.tight_layout()
            plt.savefig('original_energy_analysis.png', dpi=300, bbox_inches='tight')
            plt.show()

        # 2. Weather variables analysis
        weather_cols = ['tavg', 'tmin', 'tmax', 'prcp', 'wspd', 'pres', 'tsun']
        available_weather = [col for col in weather_cols if col in self.energy_weather_data.columns]

        if available_weather:
            n_weather = len(available_weather)
            n_cols = min(4, n_weather)
            n_rows = (n_weather + n_cols - 1) // n_cols

            fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 4*n_rows))
            fig.suptitle('Weather Variables Distribution (Original Data)', fontsize=16, fontweight='bold')

            if n_rows == 1:
                axes = [axes] if n_weather == 1 else axes
            else:
                axes = axes.flatten()

            for i, col in enumerate(available_weather):
                weather_data = self.energy_weather_data[col].dropna()
                axes[i].hist(weather_data, bins=30, alpha=0.7, edgecolor='black')
                axes[i].set_title(f'{col.upper()} Distribution')
                axes[i].set_xlabel(col)
                axes[i].set_ylabel('Frequency')

                # Add statistics text
                mean_val = weather_data.mean()
                std_val = weather_data.std()
                axes[i].axvline(mean_val, color='red', linestyle='--', alpha=0.7, label=f'Mean: {mean_val:.2f}')
                axes[i].legend()

            # Hide empty subplots
            for i in range(n_weather, len(axes)):
                axes[i].set_visible(False)

            plt.tight_layout()
            plt.savefig('original_weather_distributions.png', dpi=300, bbox_inches='tight')
            plt.show()

        # 3. Fuel type analysis
        if 'fueltype' in self.daily_fuel_data.columns and 'value' in self.daily_fuel_data.columns:
            fig, axes = plt.subplots(2, 2, figsize=(16, 12))
            fig.suptitle('Fuel Type Analysis (Original Data)', fontsize=16, fontweight='bold')

            # Fuel distribution pie chart
            fuel_totals = self.daily_fuel_data.groupby('fueltype')['value'].sum()
            axes[0, 0].pie(fuel_totals.values, labels=fuel_totals.index, autopct='%1.1f%%')
            axes[0, 0].set_title('Fuel Type Distribution')

            # Fuel distribution bar chart
            fuel_totals.plot(kind='bar', ax=axes[0, 1])
            axes[0, 1].set_title('Total Value by Fuel Type')
            axes[0, 1].set_ylabel('Total Value')
            axes[0, 1].tick_params(axis='x', rotation=45)

            # Fuel statistics
            fuel_stats = self.daily_fuel_data.groupby('fueltype')['value'].agg(['count', 'mean', 'std'])
            fuel_stats['mean'].plot(kind='bar', ax=axes[1, 0])
            axes[1, 0].set_title('Average Value by Fuel Type')
            axes[1, 0].set_ylabel('Average Value')
            axes[1, 0].tick_params(axis='x', rotation=45)

            # Fuel box plot
            fuel_types = fuel_totals.head(8).index  # Top 8 fuel types
            fuel_data_for_box = []
            fuel_labels = []
            for fuel in fuel_types:
                fuel_values = self.daily_fuel_data[self.daily_fuel_data['fueltype'] == fuel]['value']
                if len(fuel_values) > 0:
                    fuel_data_for_box.append(fuel_values)
                    fuel_labels.append(fuel)

            if fuel_data_for_box:
                axes[1, 1].boxplot(fuel_data_for_box, labels=fuel_labels)
                axes[1, 1].set_title('Value Distribution by Fuel Type (Top 8)')
                axes[1, 1].set_ylabel('Value')
                axes[1, 1].tick_params(axis='x', rotation=45)

            plt.tight_layout()
            plt.savefig('original_fuel_analysis.png', dpi=300, bbox_inches='tight')
            plt.show()

        # 4. Correlation heatmap
        if available_weather and 'value' in self.energy_weather_data.columns:
            analysis_cols = ['value'] + available_weather
            corr_data = self.energy_weather_data[analysis_cols].dropna()

            if len(corr_data) > 0:
                plt.figure(figsize=(10, 8))
                correlation_matrix = corr_data.corr()

                sns.heatmap(correlation_matrix, annot=True, cmap='RdBu_r', center=0,
                           square=True, linewidths=0.5, cbar_kws={"shrink": .8})
                plt.title('Correlation Matrix: Energy Demand vs Weather (Original Data)',
                         fontsize=14, fontweight='bold')
                plt.tight_layout()
                plt.savefig('original_correlation_heatmap.png', dpi=300, bbox_inches='tight')
                plt.show()

        print(" All visualizations for original data created and saved")

    def generate_original_data_report(self):
        """To bo co tng hp cho d liu gc"""
        print("\n" + "=" * 60)
        print("7. ORIGINAL DATA ANALYSIS SUMMARY REPORT")
        print("=" * 60)

        report = []
        report.append("ENERGY AND WEATHER DATA ANALYSIS - ORIGINAL DATA ONLY")
        report.append("=" * 55)
        report.append(f"Report generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
        report.append("")

        # Data filtering summary
        report.append("DATA FILTERING SUMMARY:")
        report.append("-" * 25)
        total_removed = len(self.removed_columns['energy_weather']) + len(self.removed_columns['daily_fuel'])
        report.append(f" Total feature engineered columns removed: {total_removed}")

        if self.removed_columns['energy_weather']:
            report.append(f" Energy Weather data - removed columns ({len(self.removed_columns['energy_weather'])}):")
            for col in self.removed_columns['energy_weather']:
                report.append(f"  - {col}")

        if self.removed_columns['daily_fuel']:
            report.append(f" Daily Fuel data - removed columns ({len(self.removed_columns['daily_fuel'])}):")
            for col in self.removed_columns['daily_fuel']:
                report.append(f"  - {col}")

        report.append("")

        # Dataset overview
        report.append("ORIGINAL DATASET OVERVIEW:")
        report.append("-" * 25)
        report.append(f" Energy Weather Data: {self.energy_weather_data.shape[0]:,} rows  {self.energy_weather_data.shape[1]} columns")
        report.append(f" Daily Fuel Data: {self.daily_fuel_data.shape[0]:,} rows  {self.daily_fuel_data.shape[1]} columns")

        # Original columns included
        report.append(f" Energy Weather original columns: {', '.join(self.energy_weather_data.columns)}")
        report.append(f" Daily Fuel original columns: {', '.join(self.daily_fuel_data.columns)}")
        report.append("")

        # Key findings from original data
        report.append("KEY FINDINGS (ORIGINAL DATA):")
        report.append("-" * 30)

        # Energy insights
        if 'value' in self.energy_weather_data.columns:
            energy_stats = self.energy_weather_data['value'].describe()
            report.append(f" Energy demand (original):")
            report.append(f"  - Mean: {energy_stats['mean']:,.2f}")
            report.append(f"  - Std: {energy_stats['std']:,.2f}")
            report.append(f"  - Range: {energy_stats['min']:,.2f} to {energy_stats['max']:,.2f}")

        # Weather insights
        weather_cols = ['tavg', 'tmin', 'tmax', 'prcp', 'wspd', 'pres', 'tsun']
        available_weather = [col for col in weather_cols if col in self.energy_weather_data.columns]
        if available_weather:
            report.append(f" Weather variables (original): {', '.join(available_weather)}")

        # Fuel insights
        if 'fueltype' in self.daily_fuel_data.columns:
            fuel_types = self.daily_fuel_data['fueltype'].nunique()
            top_fuel = self.daily_fuel_data.groupby('fueltype')['value'].sum().idxmax()
            report.append(f" Fuel types (original): {fuel_types} different types")
            report.append(f" Most used fuel type: {top_fuel}")

        # Data quality of original data
        report.append("")
        report.append("DATA QUALITY (ORIGINAL DATA):")
        report.append("-" * 30)
        energy_missing = self.energy_weather_data.isnull().sum().sum()
        fuel_missing = self.daily_fuel_data.isnull().sum().sum()
        report.append(f" Energy Weather missing values: {energy_missing}")
        report.append(f" Daily Fuel missing values: {fuel_missing}")

        energy_duplicates = self.energy_weather_data.duplicated().sum()
        fuel_duplicates = self.daily_fuel_data.duplicated().sum()
        report.append(f" Energy Weather duplicates: {energy_duplicates}")
        report.append(f" Daily Fuel duplicates: {fuel_duplicates}")

        # Recommendations
        report.append("")
        report.append("RECOMMENDATIONS FOR MODELING:")
        report.append("-" * 35)
        report.append(" Use only original features for baseline model")
        report.append(" Create new feature engineering based on domain knowledge")
        report.append(" Consider temporal features (hour, day, month, season)")
        report.append(" Analyze lag effects of weather on energy demand")
        report.append(" Investigate regional differences if multiple regions exist")

        # Save report
        full_report = "\n".join(report)
        print(full_report)

        with open('original_data_eda_report.txt', 'w', encoding='utf-8') as f:
            f.write(full_report)

        print("\n Original data EDA report saved to 'original_data_eda_report.txt'")

    def run_complete_original_eda(self, energy_weather_path, daily_fuel_path):
        """Chy EDA hon chnh cho d liu gc"""
        print(" ENERGY AND WEATHER DATA - EDA FOR ORIGINAL DATA ONLY")
        print("=" * 60)
        print(f"Analysis started: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
        print("=" * 60)

        # Load and filter data
        self.load_and_filter_data(energy_weather_path, daily_fuel_path)

        # Run all analyses on original data only
        self.basic_info_analysis()
        self.original_data_statistics()
        self.original_time_analysis()
        self.original_categorical_analysis()
        self.original_correlation_analysis()
        self.create_original_data_visualizations()
        self.generate_original_data_report()

        print("\n" + "=" * 60)
        print(" ORIGINAL DATA EDA COMPLETED SUCCESSFULLY!")
        print(f"Analysis finished: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
        print("=" * 60)

        return {
            'original_energy_weather_data': self.energy_weather_data,
            'original_daily_fuel_data': self.daily_fuel_data,
            'removed_columns': self.removed_columns
        }

# Usage
if __name__ == "__main__":
    # Initialize the Original Data EDA class
    eda = OriginalDataEDA()

    # Run complete EDA for original data only
    energy_weather_path = "/content/drive/MyDrive/DS200/Dataset/energy_weather_data_2020-06-01_to_2022-06-01_US_Center.csv"
    daily_fuel_path = "/content/drive/MyDrive/DS200/Dataset/daily_fuel_type_data_2020-06-01_to_2022-06-01.csv"

    results = eda.run_complete_original_eda(energy_weather_path, daily_fuel_path)

    print(f"\n SUMMARY:")
    print(f" Filtered to original columns only")
    print(f" Removed {len(results['removed_columns']['energy_weather'])} columns from energy_weather_data")
    print(f" Removed {len(results['removed_columns']['daily_fuel'])} columns from daily_fuel_data")
    print(f" Analysis completed on clean original data")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.express as px
from scipy import stats
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings('ignore')

# Set plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

class AZPSEnergyWeatherAnalysis:
    """
    Chuyn phn tch xu hng tiu th in v thi tit cho khu vc AZPS
    Arizona Public Service - c trng kh hu sa mc
    """

    def __init__(self):
        self.azps_data = None
        self.azps_fuel_data = None
        self.analysis_results = {}
        self.weather_cols = ['tavg', 'tmin', 'tmax', 'prcp', 'wspd', 'pres', 'tsun']

    def load_and_filter_azps_data(self, energy_weather_path, daily_fuel_path):
        """Load d liu v lc ring khu vc AZPS"""
        print("  LOADING AND FILTERING AZPS (ARIZONA PUBLIC SERVICE) DATA")
        print("=" * 60)

        # Load full datasets
        print("Loading complete datasets...")
        energy_weather_full = pd.read_csv(energy_weather_path)
        daily_fuel_full = pd.read_csv(daily_fuel_path)

        # Identify AZPS data
        print(f"Original energy_weather data: {energy_weather_full.shape}")
        print(f"Original daily_fuel data: {daily_fuel_full.shape}")

        # Filter for AZPS - multiple criteria
        azps_conditions = [
            energy_weather_full['respondent-name'].str.contains('AZPS', case=False, na=False),
            energy_weather_full['respondent-name'].str.contains('Arizona Public Service', case=False, na=False),
            energy_weather_full['respondent'].str.contains('AZPS', case=False, na=False),
            energy_weather_full['respondent-name'].str.contains('APS', case=False, na=False)
        ]

        # Combine conditions
        azps_mask = azps_conditions[0]
        for condition in azps_conditions[1:]:
            azps_mask = azps_mask | condition

        self.azps_data = energy_weather_full[azps_mask].copy()

        # Filter fuel data for AZPS
        fuel_azps_conditions = [
            daily_fuel_full['respondent-name'].str.contains('AZPS', case=False, na=False),
            daily_fuel_full['respondent-name'].str.contains('Arizona Public Service', case=False, na=False),
            daily_fuel_full['respondent'].str.contains('AZPS', case=False, na=False),
            daily_fuel_full['respondent-name'].str.contains('APS', case=False, na=False)
        ]

        fuel_azps_mask = fuel_azps_conditions[0]
        for condition in fuel_azps_conditions[1:]:
            fuel_azps_mask = fuel_azps_mask | condition

        self.azps_fuel_data = daily_fuel_full[fuel_azps_mask].copy()

        print(f"\n AZPS FILTERING RESULTS:")
        print(f"AZPS energy/weather data: {self.azps_data.shape}")
        print(f"AZPS fuel data: {self.azps_fuel_data.shape}")

        if len(self.azps_data) == 0:
            print("  WARNING: No AZPS data found with current filters")
            print("Available respondent names (sample):")
            print(energy_weather_full['respondent-name'].value_counts().head(10))
            return False

        # Show what AZPS entities were found
        print(f"\n AZPS entities found:")
        azps_entities = self.azps_data['respondent-name'].value_counts()
        for entity, count in azps_entities.items():
            print(f"   {entity}: {count:,} records")

        return True

    def prepare_azps_time_series(self):
        """Chun b d liu time series cho AZPS"""
        print(f"\n PREPARING AZPS TIME SERIES DATA")
        print("-" * 40)

        if self.azps_data is None or len(self.azps_data) == 0:
            print(" No AZPS data available")
            return False

        # Convert time columns
        for time_col in ['period', 'time']:
            if time_col in self.azps_data.columns:
                self.azps_data[time_col] = pd.to_datetime(self.azps_data[time_col], errors='coerce')

        if 'period' in self.azps_fuel_data.columns:
            self.azps_fuel_data['period'] = pd.to_datetime(self.azps_fuel_data['period'], errors='coerce')

        # Create time-based features
        if 'period' in self.azps_data.columns:
            df = self.azps_data
            df['year'] = df['period'].dt.year
            df['month'] = df['period'].dt.month
            df['day'] = df['period'].dt.day
            df['hour'] = df['period'].dt.hour
            df['dayofweek'] = df['period'].dt.dayofweek  # 0=Monday, 6=Sunday
            df['dayofyear'] = df['period'].dt.dayofyear
            df['week'] = df['period'].dt.isocalendar().week

            # Arizona seasons (desert climate)
            def arizona_season(month):
                if month in [12, 1, 2]:
                    return 'Winter'
                elif month in [3, 4, 5]:
                    return 'Spring'
                elif month in [6, 7, 8]:
                    return 'Summer'
                else:  # 9, 10, 11
                    return 'Fall'

            df['season'] = df['month'].apply(arizona_season)

            # Arizona specific periods
            df['is_summer_peak'] = ((df['month'] >= 6) & (df['month'] <= 8) &
                                   (df['hour'] >= 15) & (df['hour'] <= 19))  # Peak cooling hours
            df['is_monsoon'] = (df['month'] >= 7) & (df['month'] <= 9)  # Monsoon season
            df['is_weekend'] = df['dayofweek'] >= 5

        # Time range analysis
        time_cols = [col for col in ['period', 'time'] if col in self.azps_data.columns]
        for col in time_cols:
            time_data = self.azps_data[col].dropna()
            if len(time_data) > 0:
                print(f"{col}: {time_data.min()} to {time_data.max()}")
                print(f"   Duration: {time_data.max() - time_data.min()}")
                print(f"   Records: {len(time_data):,}")

        return True

    def azps_consumption_patterns(self):
        """Phn tch patterns tiu th in ca AZPS"""
        print(f"\n AZPS ELECTRICITY CONSUMPTION PATTERNS")
        print("=" * 45)

        if 'value' not in self.azps_data.columns or 'type' not in self.azps_data.columns:
            print(" No consumption data (value column) or type column found")
            return

        # Filter for 'Demand' type
        demand_data = self.azps_data[self.azps_data['type'] == 'D'].copy()

        if len(demand_data) == 0:
            print(" No 'Demand' data found for AZPS")
            return

        consumption = demand_data['value'].dropna()

        # Basic statistics
        print(f" AZPS Consumption Statistics (Demand Type Only):")
        print(f"   Records: {len(consumption):,}")
        print(f"   Mean: {consumption.mean():,.2f} {demand_data['value-units'].iloc[0] if 'value-units' in demand_data.columns else 'units'}")
        print(f"   Median: {consumption.median():,.2f}")
        print(f"   Std: {consumption.std():,.2f}")
        print(f"   Min: {consumption.min():,.2f}")
        print(f"   Max: {consumption.max():,.2f}")
        # Handle division by zero if mean is zero or close to zero
        peak_avg_ratio = consumption.max()/consumption.mean() if consumption.mean() != 0 else float('inf')
        print(f"   Peak/Average Ratio: {peak_avg_ratio:.2f}")

        # Load factor (efficiency metric)
        load_factor = consumption.mean() / consumption.max() * 100 if consumption.max() != 0 else 0
        print(f"   Load Factor: {load_factor:.1f}%")

        # Time-based patterns
        if 'hour' in demand_data.columns:
            print(f"\n Hourly Consumption Patterns (Demand Type Only):")
            hourly_avg = demand_data.groupby('hour')['value'].mean()
            peak_hour = hourly_avg.idxmax()
            min_hour = hourly_avg.idxmin()
            print(f"   Peak hour: {peak_hour}:00 ({hourly_avg[peak_hour]:,.2f})")
            print(f"   Min hour: {min_hour}:00 ({hourly_avg[min_hour]:,.2f})")
            # Handle division by zero
            peak_min_ratio = hourly_avg[peak_hour]/hourly_avg[min_hour] if hourly_avg[min_hour] != 0 else float('inf')
            print(f"   Peak/Min ratio: {peak_min_ratio:.2f}")


        if 'season' in demand_data.columns:
            print(f"\n  Seasonal Consumption (Arizona Climate) (Demand Type Only):")
            seasonal_stats = demand_data.groupby('season')['value'].agg(['mean', 'max', 'std']).round(2)
            for season in ['Winter', 'Spring', 'Summer', 'Fall']:
                if season in seasonal_stats.index:
                    stats = seasonal_stats.loc[season]
                    print(f"   {season}: Mean={stats['mean']:,.0f}, Max={stats['max']:,.0f}, Std={stats['std']:,.0f}")

        if 'is_weekend' in demand_data.columns:
            print(f"\n Weekday vs Weekend (Demand Type Only):")
            weekday_avg = demand_data[~demand_data['is_weekend']]['value'].mean()
            weekend_avg = demand_data[demand_data['is_weekend']]['value'].mean()
            print(f"   Weekday average: {weekday_avg:,.2f}")
            print(f"   Weekend average: {weekend_avg:,.2f}")
            # Handle division by zero
            weekend_weekday_ratio = weekend_avg/weekday_avg if weekday_avg != 0 else float('inf')
            print(f"   Weekend/Weekday ratio: {weekend_weekday_ratio:.3f}")


        self.analysis_results['consumption_stats'] = {
            'mean': consumption.mean(),
            'peak': consumption.max(),
            'load_factor': load_factor,
            'peak_hour': peak_hour if 'hour' in demand_data.columns else None
        }


    def azps_weather_impact_analysis(self):
        """Phn tch tc ng thi tit n tiu th in AZPS"""
        print(f"\n  AZPS WEATHER IMPACT ANALYSIS")
        print("=" * 35)

        if 'type' not in self.azps_data.columns:
            print(" 'type' column not found in AZPS data")
            return

        # Filter for 'Demand' type
        demand_data = self.azps_data[self.azps_data['type'] == 'D'].copy()

        if len(demand_data) == 0:
            print(" No 'Demand' data found for AZPS")
            return

        available_weather = [col for col in self.weather_cols if col in demand_data.columns]

        if not available_weather or 'value' not in demand_data.columns:
            print(" Insufficient weather or consumption data after filtering for 'Demand'")
            return

        print(f"Available weather variables for Demand data: {', '.join(available_weather)}")

        # Weather-consumption correlations
        print(f"\n Weather-Consumption Correlations (Demand Type Only):")
        weather_corr = {}
        for weather_var in available_weather:
            clean_data = demand_data[[weather_var, 'value']].dropna()
            if len(clean_data) > 10:
                corr = clean_data[weather_var].corr(clean_data['value'])
                weather_corr[weather_var] = corr
                strength = "Strong" if abs(corr) > 0.7 else "Moderate" if abs(corr) > 0.3 else "Weak"
                print(f"   {weather_var:>6}: {corr:>7.3f} ({strength})")

        # Temperature analysis (key for Arizona)
        if 'tavg' in available_weather:
            print(f"\n  Temperature-Consumption Analysis (Arizona Desert Climate) (Demand Type Only):")
            temp_data = demand_data[['tavg', 'value']].dropna()

            if len(temp_data) > 0:
                # Temperature bins analysis
                temp_data['temp_bin'] = pd.cut(temp_data['tavg'], bins=10)
                temp_consumption = temp_data.groupby('temp_bin')['value'].mean()

                print(f"   Temperature range: {temp_data['tavg'].min():.1f}C to {temp_data['tavg'].max():.1f}C")
                print(f"   Consumption at coolest temps: {temp_consumption.iloc[0]:,.0f}")
                print(f"   Consumption at hottest temps: {temp_consumption.iloc[-1]:,.0f}")

                # Cooling/Heating degree days approximation
                # Using 10th percentile of tavg for Demand data as base temperature
                base_temp = temp_data['tavg'].quantile(0.1)
                temp_data['cdd'] = np.maximum(temp_data['tavg'] - base_temp, 0)  # Cooling degree days
                temp_data['hdd'] = np.maximum(base_temp - temp_data['tavg'], 0)  # Heating degree days

                cdd_corr = temp_data['cdd'].corr(temp_data['value'])
                hdd_corr = temp_data['hdd'].corr(temp_data['value'])

                print(f"   Base temperature (approx, based on 10th percentile of Demand tavg): {base_temp:.1f}C")
                print(f"   Cooling Degree Days correlation: {cdd_corr:.3f}")
                print(f"   Heating Degree Days correlation: {hdd_corr:.3f}")

                if abs(cdd_corr) > abs(hdd_corr):
                    print(f"    COOLING DOMINATED (typical for Arizona)")
                else:
                    print(f"    HEATING DOMINATED (unusual for Arizona)")

        # Seasonal weather patterns
        if 'season' in demand_data.columns and available_weather:
            print(f"\n  Seasonal Weather-Consumption Patterns (Demand Type Only):")
            for season in ['Winter', 'Spring', 'Summer', 'Fall']:
                season_data = demand_data[demand_data['season'] == season]
                if len(season_data) > 10:
                    season_consumption = season_data['value'].mean()
                    season_temp = season_data['tavg'].mean() if 'tavg' in season_data.columns else None
                    print(f"   {season:>6}: Consumption={season_consumption:,.0f}", end="")
                    if season_temp is not None:
                        print(f", Avg Temp={season_temp:.1f}C")
                    else:
                        print()

        # Extreme weather analysis
        if 'tavg' in available_weather:
            print(f"\n Extreme Weather Impact (Arizona Heat Waves) (Demand Type Only):")
            temp_95th = demand_data['tavg'].quantile(0.95)
            temp_5th = demand_data['tavg'].quantile(0.05)

            extreme_hot_data = demand_data[demand_data['tavg'] >= temp_95th]
            extreme_cold_data = demand_data[demand_data['tavg'] <= temp_5th]
            normal_data = demand_data[(demand_data['tavg'] > temp_5th) & (demand_data['tavg'] < temp_95th)]

            extreme_hot = extreme_hot_data['value'].mean() if len(extreme_hot_data) > 0 else np.nan
            extreme_cold = extreme_cold_data['value'].mean() if len(extreme_cold_data) > 0 else np.nan
            normal_consumption = normal_data['value'].mean() if len(normal_data) > 0 else np.nan

            print(f"   Extreme hot days ({temp_95th:.1f}C): {extreme_hot:,.0f} consumption")
            print(f"   Extreme cold days ({temp_5th:.1f}C): {extreme_cold:,.0f} consumption")
            print(f"   Normal conditions: {normal_consumption:,.0f} consumption")
            # Calculate percentage change, handling potential division by zero or NaN
            hot_impact_pct = ((extreme_hot / normal_consumption - 1) * 100) if not np.isnan(extreme_hot) and not np.isnan(normal_consumption) and normal_consumption != 0 else np.nan
            cold_impact_pct = ((extreme_cold / normal_consumption - 1) * 100) if not np.isnan(extreme_cold) and not np.isnan(normal_consumption) and normal_consumption != 0 else np.nan
            print(f"   Hot weather impact: {hot_impact_pct:+.1f}%" if not np.isnan(hot_impact_pct) else "   Hot weather impact: N/A")
            print(f"   Cold weather impact: {cold_impact_pct:+.1f}%" if not np.isnan(cold_impact_pct) else "   Cold weather impact: N/A")


        self.analysis_results['weather_impact'] = weather_corr

    def azps_solar_potential_analysis(self):
        """Phn tch tim nng solar cho AZPS (Arizona c tim nng solar cao)"""
        print(f"\n  AZPS SOLAR POTENTIAL ANALYSIS")
        print("=" * 30)

        if 'tsun' not in self.azps_data.columns:
            print(" Solar data (tsun) not available")
            return

        solar_data = self.azps_data['tsun'].dropna()

        if len(solar_data) == 0:
            print(" No valid solar data")
            return

        print(f" Solar Irradiance Statistics (Arizona Desert):")
        print(f"   Mean sunshine hours: {solar_data.mean():.2f} hours/day")
        print(f"   Max sunshine hours: {solar_data.max():.2f} hours/day")
        print(f"   Min sunshine hours: {solar_data.min():.2f} hours/day")
        print(f"   Days with >10h sunshine: {(solar_data > 10).sum()} ({(solar_data > 10).mean()*100:.1f}%)")

        # Solar vs consumption relationship (using Demand data)
        if 'value' in self.azps_data.columns and 'type' in self.azps_data.columns:
            demand_data = self.azps_data[self.azps_data['type'] == 'D'].copy()
            if len(demand_data) > 0 and 'tsun' in demand_data.columns:
                solar_consumption_data = demand_data[['tsun', 'value']].dropna()
                if len(solar_consumption_data) > 10:
                    solar_corr = solar_consumption_data['tsun'].corr(solar_consumption_data['value'])
                    print(f"\n Solar-Consumption Correlation (Demand Type Only): {solar_corr:.3f}")

                    if solar_corr < -0.3:
                        print(f"    Negative correlation: More sun = Less consumption (due to solar generation offsetting demand)")
                    elif solar_corr > 0.3:
                        print(f"    Positive correlation: More sun = More consumption (likely due to increased AC load on sunny days)")
                    else:
                        print(f"    Weak correlation: Solar irradiance has minimal direct correlation with demand")

        # Seasonal solar patterns
        if 'season' in self.azps_data.columns:
            print(f"\n  Seasonal Solar Patterns:")
            seasonal_solar = self.azps_data.groupby('season')['tsun'].mean()
            for season in ['Winter', 'Spring', 'Summer', 'Fall']:
                if season in seasonal_solar.index:
                    print(f"   {season:>6}: {seasonal_solar[season]:.2f} hours/day average")


    def create_azps_visualizations(self):
        """To visualizations chuyn bit cho AZPS"""
        print(f"\n CREATING AZPS-SPECIFIC VISUALIZATIONS")
        print("=" * 40)

        # Filter data for 'Demand' type for most plots
        if 'type' not in self.azps_data.columns:
             print("  'type' column not found. Cannot filter for Demand.")
             demand_plot_data = self.azps_data.copy() # Use full data if filter not possible
        else:
             demand_plot_data = self.azps_data[self.azps_data['type'] == 'D'].copy()

        if len(demand_plot_data) == 0:
             print("  No 'Demand' data found for plotting.")
             return # Exit if no demand data


        # 1. AZPS Consumption Time Series (Demand Only)
        if 'period' in demand_plot_data.columns and 'value' in demand_plot_data.columns:
            plt.figure(figsize=(16, 12))

            # Main time series
            plt.subplot(3, 2, 1)
            time_sorted = demand_plot_data.sort_values('period')
            plt.plot(time_sorted['period'], time_sorted['value'], alpha=0.7, linewidth=0.8)
            plt.title('AZPS Electricity Demand Over Time')
            plt.xlabel('Time')
            plt.ylabel('Demand (MWh)')
            plt.xticks(rotation=45)

            # Seasonal patterns (Demand Only)
            if 'season' in demand_plot_data.columns:
                plt.subplot(3, 2, 2)
                seasonal_data = demand_plot_data.groupby('season')['value'].mean()
                seasonal_data.plot(kind='bar', color=['skyblue', 'lightgreen', 'red', 'orange'])
                plt.title('Average Demand by Season (Arizona)')
                plt.ylabel('Average Demand (MWh)')
                plt.xticks(rotation=45)

            # Hourly patterns (Demand Only)
            if 'hour' in demand_plot_data.columns:
                plt.subplot(3, 2, 3)
                hourly_data = demand_plot_data.groupby('hour')['value'].mean()
                plt.plot(hourly_data.index, hourly_data.values, marker='o')
                plt.title('AZPS Daily Load Profile (Average Demand)')
                plt.xlabel('Hour of Day')
                plt.ylabel('Average Demand (MWh)')
                plt.grid(True, alpha=0.3)

            # Weekend vs Weekday (Demand Only)
            if 'is_weekend' in demand_plot_data.columns:
                plt.subplot(3, 2, 4)
                weekend_comparison = demand_plot_data.groupby('is_weekend')['value'].mean()
                labels = ['Weekday', 'Weekend']
                plt.bar(labels, weekend_comparison.values, color=['blue', 'orange'])
                plt.title('Weekday vs Weekend Demand (Average)')
                plt.ylabel('Average Demand (MWh)')

            # Temperature vs Consumption scatter (Demand Only)
            if 'tavg' in demand_plot_data.columns:
                plt.subplot(3, 2, 5)
                temp_cons_data = demand_plot_data[['tavg', 'value']].dropna()
                plt.scatter(temp_cons_data['tavg'], temp_cons_data['value'], alpha=0.5, s=1)
                plt.title('Temperature vs Demand (Arizona Desert)')
                plt.xlabel('Average Temperature (C)')
                plt.ylabel('Demand (MWh)')

                # Add trend line
                if len(temp_cons_data) > 10:
                    z = np.polyfit(temp_cons_data['tavg'], temp_cons_data['value'], 1)
                    p = np.poly1d(z)
                    plt.plot(temp_cons_data['tavg'], p(temp_cons_data['tavg']), "r--", alpha=0.8)

            # Monthly heatmap (Demand Only)
            if 'month' in demand_plot_data.columns and 'hour' in demand_plot_data.columns:
                plt.subplot(3, 2, 6)
                monthly_hourly = demand_plot_data.groupby(['month', 'hour'])['value'].mean().unstack()
                im = plt.imshow(monthly_hourly.values, cmap='YlOrRd', aspect='auto')
                plt.colorbar(im)
                plt.title('Demand Heatmap: Month vs Hour')
                plt.xlabel('Hour of Day')
                plt.ylabel('Month')
                plt.yticks(range(12), monthly_hourly.index) # Use 0-indexed months for plotting

            plt.tight_layout()
            plt.savefig('azps_consumption_analysis.png', dpi=300, bbox_inches='tight')
            plt.show()

        # 2. Weather Impact Visualization (Demand Only)
        available_weather_demand = [col for col in self.weather_cols if col in demand_plot_data.columns]
        if available_weather_demand and 'value' in demand_plot_data.columns:
            n_weather = len(available_weather_demand)
            n_cols = min(3, n_weather)
            n_rows = (n_weather + n_cols - 1) // n_cols

            plt.figure(figsize=(15, 5*n_rows))
            plt.suptitle('AZPS Weather-Demand Relationships', fontsize=16, fontweight='bold')

            for i, weather_var in enumerate(available_weather_demand):
                plt.subplot(n_rows, n_cols, i+1)
                weather_data = demand_plot_data[[weather_var, 'value']].dropna()

                if len(weather_data) > 10:
                    plt.scatter(weather_data[weather_var], weather_data['value'], alpha=0.5, s=1)

                    # Correlation
                    corr = weather_data[weather_var].corr(weather_data['value'])
                    plt.title(f'{weather_var.upper()} vs Demand\nCorr: {corr:.3f}')
                    plt.xlabel(weather_var)
                    plt.ylabel('Demand (MWh)')

                    # Trend line
                    z = np.polyfit(weather_data[weather_var], weather_data['value'], 1)
                    p = np.poly1d(z)
                    plt.plot(weather_data[weather_var], p(weather_data[weather_var]), "r--", alpha=0.8)

            plt.tight_layout()
            plt.savefig('azps_weather_impact.png', dpi=300, bbox_inches='tight')
            plt.show()

        # 3. Interactive Dashboard (Demand Only)
        self.create_azps_interactive_dashboard(demand_plot_data) # Pass filtered data

        print(" All AZPS visualizations created and saved")

    def create_azps_interactive_dashboard(self, df_demand):
        """To interactive dashboard cho AZPS (Demand Only)"""
        if df_demand is None or len(df_demand) == 0 or 'period' not in df_demand.columns or 'value' not in df_demand.columns:
            print("  Cannot create interactive dashboard - insufficient Demand data")
            return

        # Create subplots
        fig = make_subplots(
            rows=2, cols=2,
            subplot_titles=('AZPS Demand Timeline', 'Temperature vs Demand',
                          'Seasonal Patterns (Demand)', 'Daily Load Profile (Demand)'),
            specs=[[{"secondary_y": False}, {"secondary_y": False}],
                   [{"secondary_y": False}, {"secondary_y": False}]]
        )

        # 1. Timeline
        time_sorted = df_demand.sort_values('period')
        fig.add_trace(
            go.Scatter(x=time_sorted['period'], y=time_sorted['value'],
                      mode='lines', name='AZPS Demand', line=dict(width=1)),
            row=1, col=1
        )

        # 2. Temperature scatter (if available)
        if 'tavg' in df_demand.columns:
            temp_data = df_demand[['tavg', 'value']].dropna()
            fig.add_trace(
                go.Scatter(x=temp_data['tavg'], y=temp_data['value'],
                          mode='markers', name='Temp vs Demand',
                          marker=dict(size=3, opacity=0.6)),
                row=1, col=2
            )

        # 3. Seasonal patterns
        if 'season' in df_demand.columns:
            seasonal_avg = df_demand.groupby('season')['value'].mean()
            fig.add_trace(
                go.Bar(x=seasonal_avg.index, y=seasonal_avg.values,
                       name='Seasonal Average'),
                row=2, col=1
            )

        # 4. Daily profile
        if 'hour' in df_demand.columns:
            hourly_avg = df_demand.groupby('hour')['value'].mean()
            fig.add_trace(
                go.Scatter(x=hourly_avg.index, y=hourly_avg.values,
                          mode='lines+markers', name='Daily Profile'),
                row=2, col=2
            )

        fig.update_layout(height=800, showlegend=True,
                         title_text="AZPS Energy Demand Dashboard")

        fig.write_html("azps_energy_demand_dashboard.html") # Updated filename
        print(" Interactive dashboard saved as 'azps_energy_demand_dashboard.html'")


    def azps_fuel_mix_analysis(self):
        """Phn tch fuel mix ca AZPS"""
        print(f"\n AZPS FUEL MIX ANALYSIS")
        print("=" * 25)

        if self.azps_fuel_data is None or len(self.azps_fuel_data) == 0:
            print(" No AZPS fuel data available")
            return

        print(f"AZPS Fuel Data: {self.azps_fuel_data.shape}")

        if 'fueltype' in self.azps_fuel_data.columns and 'value' in self.azps_fuel_data.columns:
            fuel_totals = self.azps_fuel_data.groupby('fueltype')['value'].sum().sort_values(ascending=False)
            fuel_percentages = (fuel_totals / fuel_totals.sum() * 100).round(2)

            print(f"\n AZPS Fuel Mix (by total generation):")
            for fuel, percentage in fuel_percentages.items():
                print(f"   {fuel}: {percentage}% ({fuel_totals[fuel]:,.0f} total)")

            # Arizona-specific fuel analysis (using fueltype strings)
            renewable_fuels_keywords = ['solar', 'wind', 'hydro', 'geothermal']
            fossil_fuels_keywords = ['natural gas', 'coal', 'petroleum']

            renewable_total_pct = 0
            fossil_total_pct = 0
            other_nuclear_total_pct = 0

            for fuel_type, percentage in fuel_percentages.items():
                fuel_type_lower = fuel_type.lower()
                if any(keyword in fuel_type_lower for keyword in renewable_fuels_keywords):
                    renewable_total_pct += percentage
                elif any(keyword in fuel_type_lower for keyword in fossil_fuels_keywords):
                    fossil_total_pct += percentage
                else:
                    other_nuclear_total_pct += percentage # Assuming remaining are Other or Nuclear

            print(f"\n Renewable vs Fossil Fuel Mix (Estimated):")
            print(f"   Estimated Renewable: {renewable_total_pct:.1f}%")
            print(f"   Estimated Fossil: {fossil_total_pct:.1f}%")
            print(f"   Other/Nuclear: {other_nuclear_total_pct:.1f}%")


    def generate_azps_report(self):
        """To bo co tng hp cho AZPS"""
        print(f"\n GENERATING COMPREHENSIVE AZPS REPORT")
        print("=" * 40)

        report = []
        report.append("ARIZONA PUBLIC SERVICE (AZPS) - ENERGY DEMAND ANALYSIS")
        report.append("=" * 65)
        report.append(f"Report generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')} UTC")
        report.append(f"Analysis Focus: Desert Climate Energy Demand Patterns")
        report.append("")

        # Executive Summary
        report.append("EXECUTIVE SUMMARY")
        report.append("-" * 17)

        # Filter for 'Demand' data for report summary
        if self.azps_data is not None and 'type' in self.azps_data.columns:
            demand_data_for_report = self.azps_data[self.azps_data['type'] == 'D'].copy()
            if len(demand_data_for_report) > 0:
                consumption_mean = demand_data_for_report['value'].mean()
                consumption_peak = demand_data_for_report['value'].max()
                data_span = f"{demand_data_for_report['period'].min()} to {demand_data_for_report['period'].max()}"
                load_factor = self.analysis_results.get('consumption_stats', {}).get('load_factor', 'N/A')


                report.append(f" Dataset: {len(demand_data_for_report):,} demand records")
                report.append(f" Time Period: {data_span}")
                report.append(f" Average Demand: {consumption_mean:,.2f}")
                report.append(f" Peak Demand: {consumption_peak:,.2f}")
                report.append(f" Load Factor: {load_factor:.1f}%" if isinstance(load_factor, (int, float)) else f" Load Factor: {load_factor}")

        # Key Findings
        report.append("")
        report.append("KEY FINDINGS - ARIZONA DESERT CLIMATE IMPACTS ON DEMAND")
        report.append("-" * 55)

        if 'weather_impact' in self.analysis_results:
            weather_corrs = self.analysis_results['weather_impact']
            strongest_corr = max(weather_corrs.items(), key=lambda x: abs(x[1])) if weather_corrs else None

            if strongest_corr:
                var, corr = strongest_corr
                report.append(f" Strongest weather correlation with Demand: {var} ({corr:.3f})")

                if var == 'tavg':
                     if corr > 0.5:
                         report.append("   High temperature strongly drives cooling demand (typical Arizona pattern)")
                     elif corr < -0.5:
                         report.append("   Temperature negatively correlated (unusual for Arizona demand)")
                     else:
                         report.append("   Temperature moderately correlated with demand")


        # Seasonal Insights (Demand Only)
        if 'season' in self.azps_data.columns and 'type' in self.azps_data.columns:
            demand_data_seasonal = self.azps_data[self.azps_data['type'] == 'D'].copy()
            if len(demand_data_seasonal) > 0:
                report.append("")
                seasonal_consumption = demand_data_seasonal.groupby('season')['value'].mean()

                # Handle cases where a season might be missing or have zero demand
                peak_season = seasonal_consumption.idxmax() if not seasonal_consumption.empty else 'N/A'
                low_season = seasonal_consumption.idxmin() if not seasonal_consumption.empty else 'N/A'
                peak_val = seasonal_consumption.max() if not seasonal_consumption.empty else np.nan
                low_val = seasonal_consumption.min() if not seasonal_consumption.empty else np.nan


                report.append(f" Peak demand season: {peak_season} ({peak_val:,.0f})" if not np.isnan(peak_val) else f" Peak demand season: {peak_season} (N/A)")
                report.append(f" Lowest demand season: {low_season} ({low_val:,.0f})" if not np.isnan(low_val) else f" Lowest demand season: {low_season} (N/A)")
                # Handle division by zero or NaN for seasonal variation
                seasonal_variation_pct = ((peak_val / low_val - 1) * 100) if not np.isnan(peak_val) and not np.isnan(low_val) and low_val != 0 else np.nan
                report.append(f" Seasonal variation in demand: {seasonal_variation_pct:.1f}%" if not np.isnan(seasonal_variation_pct) else " Seasonal variation in demand: N/A")

        # Peak Load Analysis (Demand Only)
        if 'is_summer_peak' in self.azps_data.columns and 'type' in self.azps_data.columns:
            demand_data_peak = self.azps_data[self.azps_data['type'] == 'D'].copy()
            if len(demand_data_peak) > 0:
                summer_peak_consumption_data = demand_data_peak[demand_data_peak['is_summer_peak']]
                normal_consumption_data = demand_data_peak[~demand_data_peak['is_summer_peak']]

                summer_peak_consumption = summer_peak_consumption_data['value'].mean() if len(summer_peak_consumption_data) > 0 else np.nan
                normal_consumption = normal_consumption_data['value'].mean() if len(normal_consumption_data) > 0 else np.nan

                # Handle division by zero or NaN for peak premium
                peak_premium = ((summer_peak_consumption / normal_consumption - 1) * 100) if not np.isnan(summer_peak_consumption) and not np.isnan(normal_consumption) and normal_consumption != 0 else np.nan

                report.append("")
                report.append(f" Summer peak hours (3-7 PM) average demand: {summer_peak_consumption:,.0f}" if not np.isnan(summer_peak_consumption) else " Summer peak hours (3-7 PM) average demand: N/A")
                report.append(f" Off-peak average demand: {normal_consumption:,.0f}" if not np.isnan(normal_consumption) else " Off-peak average demand: N/A")
                report.append(f" Peak demand premium: {peak_premium:+.1f}%" if not np.isnan(peak_premium) else " Peak demand premium: N/A")


        # Recommendations
        report.append("")
        report.append("RECOMMENDATIONS FOR AZPS OPERATIONS")
        report.append("-" * 35)
        report.append(" Demand Response Programs: Target summer peak hours (3-7 PM)")
        report.append(" Solar Integration: Leverage Arizona's excellent solar resource")
        report.append(" Energy Storage: Address duck curve with battery storage")
        report.append(" Cooling Efficiency: Promote efficient AC systems and insulation")
        report.append(" Peak Shaving: Time-of-use rates for residential customers")

        # Data Quality Assessment (using Demand data)
        report.append("")
        report.append("DATA QUALITY ASSESSMENT (Demand Data)")
        report.append("-" * 35)
        if self.azps_data is not None and 'type' in self.azps_data.columns:
            demand_data_quality = self.azps_data[self.azps_data['type'] == 'D'].copy()
            if len(demand_data_quality) > 0:
                missing_pct = (demand_data_quality.isnull().sum().sum() / demand_data_quality.size) * 100 if demand_data_quality.size > 0 else 0
                report.append(f" Overall missing data: {missing_pct:.2f}%")

                if 'value' in demand_data_quality.columns:
                    consumption_missing = demand_data_quality['value'].isnull().sum()
                    report.append(f" Missing demand records: {consumption_missing}")

                weather_available = [col for col in self.weather_cols if col in demand_data_quality.columns]
                report.append(f" Available weather variables: {len(weather_available)}/{len(self.weather_cols)}")
            else:
                report.append(" No Demand data available for quality assessment.")


        # Technical Notes
        report.append("")
        report.append("TECHNICAL NOTES")
        report.append("-" * 15)
        report.append(" Time Zone: Arizona does not observe Daylight Saving Time")
        report.append(" Climate: Hot desert climate with extreme summer temperatures")
        report.append(" Peak Season: Summer (June-August) due to cooling loads")
        report.append(" Monsoon Season: July-September affects solar generation")

        # Save report
        full_report = "\n".join(report)
        print(full_report)

        with open('azps_energy_demand_analysis_report.txt', 'w', encoding='utf-8') as f: # Updated filename
            f.write(full_report)

        print("\n AZPS demand analysis report saved to 'azps_energy_demand_analysis_report.txt'")


    def run_complete_azps_analysis(self, energy_weather_path, daily_fuel_path):
        """Chy phn tch hon chnh cho AZPS"""
        print("  ARIZONA PUBLIC SERVICE (AZPS) - COMPREHENSIVE ENERGY DEMAND ANALYSIS")
        print("=" * 70)
        print(f"Analysis started: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')} UTC")
        print("Focus: Desert climate energy demand patterns")
        print("=" * 70)

        # Load and filter AZPS data
        if not self.load_and_filter_azps_data(energy_weather_path, daily_fuel_path):
            print(" Failed to load AZPS data. Analysis terminated.")
            return None

        # Prepare time series
        if not self.prepare_azps_time_series():
            print(" Failed to prepare time series. Analysis terminated.")
            return None

        # Run all AZPS-specific analyses
        self.azps_consumption_patterns() # Now filtered for Demand
        self.azps_weather_impact_analysis() # Now filtered for Demand
        self.azps_solar_potential_analysis() # Now considers Demand correlation
        self.azps_fuel_mix_analysis()
        self.create_azps_visualizations() # Now filtered for Demand where applicable
        self.generate_azps_report() # Now focused on Demand

        print("\n" + "=" * 70)
        print(" AZPS DEMAND ANALYSIS COMPLETED SUCCESSFULLY!")
        print(f"Analysis finished: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')} UTC")
        print("=" * 70)
        print(" Generated Files:")
        print("    azps_consumption_analysis.png (Now shows Demand)")
        print("     azps_weather_impact.png (Now shows Demand)")
        print("    azps_energy_demand_dashboard.html") # Updated filename
        print("    azps_energy_demand_analysis_report.txt") # Updated filename
        print("=" * 70)


        return {
            'azps_data': self.azps_data, # Original AZPS data with added time features
            'azps_fuel_data': self.azps_fuel_data,
            'analysis_results': self.analysis_results
        }

# Usage example
if __name__ == "__main__":
    # Initialize AZPS analyzer
    azps_analyzer = AZPSEnergyWeatherAnalysis()

    # File paths - update with your actual file locations
    energy_weather_path = "/content/drive/MyDrive/DS200/Dataset/energy_weather_data_2020-06-01_to_2022-06-01_US_Center.csv"
    daily_fuel_path = "/content/drive/MyDrive/DS200/Dataset/daily_fuel_type_data_2020-06-01_to_2022-06-01.csv"

    # Run complete AZPS analysis
    results = azps_analyzer.run_complete_azps_analysis(energy_weather_path, daily_fuel_path)

    if results:
        print(f"\n AZPS ANALYSIS SUMMARY:")
        print(f" AZPS energy data (filtered for AZPS entities): {results['azps_data'].shape}")
        print(f" AZPS fuel data (filtered for AZPS entities): {results['azps_fuel_data'].shape}")
        print(f" Key insights extracted for Arizona desert climate patterns based on Demand")
```
