```python
# Lấy model, scaler, feature tương ứng respondent AEC
import numpy as np

features=data['AEC']['features']
scaler=data['AEC']['scaler']
model=data['AEC']['model']

# Loại 'value' khỏi feature


# Lấy dữ liệu sample
a = df[(df['date'] == "2020-09-20") & (df['respondent'] == 'AEC')].copy()
print(a)
# Chắc chắn chỉ giữ đúng các cột cần thiết
# Giữ nguyên selected_features bao gồm 'value'
test = a[features].copy()  # Gồm cả 'value'
target = a['value'].copy()

# Scale đúng dữ liệu
test_scaled = scaler.transform(test)

# Sau khi scale xong, bạn loại cột 'value' nếu cần
# Giả sử 'value' là cột đầu tiên
arr_new = test_scaled[:, 1:]

# Reshape cho model
input_for_model = arr_new[np.newaxis, :, :]  # (1, 1, num_features)

# Dự đoán
prediction = model.predict(input_for_model).flatten()[0]
print(f"Dự báo giá trị: {prediction}")
print(f'Dữ liệu thật:{test_scaled[0][0]}')
```
